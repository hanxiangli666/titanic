# 资料科学框架

1. **问题是什么？**  
   如果资料科学、大数据、机器学习、预测分析、商业智慧，或任何热门 buzzword 是“解方”，  
   那么“问题”是什么？  
   老话一句：**不要本末倒置。**  
   问题 → 需求 → 解决方案 → 设计 → 技术。  
   我们常常太快跳到新的科技、工具或算法，却还没搞清楚我们要解的是什么问题。

2. **资料集在哪里？**  
   John Naisbitt 在 1984 年的《Megatrends》中写道：  
   **“我们淹没在资料之中，却饥渴于知识。”**  
   因此资料集很可能已经存在，只是散落在不同的格式或来源：  
   - 外部或内部  
   - 结构化或非结构化  
   - 静态或串流  
   - 客观或主观  
   你不需要重新发明轮子，只需要知道去哪里找。  
   下一步我们会将“脏资料”变成“干净资料”。

3. **准备可使用的资料。**  
   这个步骤通常称为 **资料整理（Data Wrangling）**：  
   将“野生”资料变成“可管理”资料。  
   资料整理包含：  
   - 建立资料储存与处理架构  
   - 建立资料治理与品质标准  
   - 资料撷取（ETL、爬虫）  
   - 资料清理：找出异常值、缺失值、离群值  

4. **进行探索性分析（EDA）。**  
   所有人都知道：**Garbage In, Garbage Out（垃圾进、垃圾出）**。  
   因此需要使用描述性统计、图形统计来找出：  
   - 潜在问题  
   - 模式  
   - 分类  
   - 相关性  
   - 比较  
   同时，理解资料型态（类别型 vs 连续型）也很重要，  
   这会影响后续使用的假设检定或模型类型。

5. **建立模型。**  
   就像描述统计与推论统计一样，资料模型可以用来：  
   - 总结资料  
   - 预测未来结果  
   算法不是魔法棒，而是工具。  
   如果你说你需要十字螺丝起子，而对方给你一支一字起子甚至是铁锤，  
   最好的情况是不专业，最糟的是无法完成任务。  
   资料建模也是如此：选择合适的工具非常重要。  
   **你才是工匠，而算法只是工具。**

6. **验证并实作模型。**  
   在使用训练集建立模型后，必须用测试集进行验证。  
   这是为了避免：  
   - 过度拟合（overfitting）  
   - 使模型只适用于特定资料，而无法泛化到同一资料集的其他子集  

7. **优化与制定策略。**  
   这是“六百万美元人（The Bionic Man）”的步骤：  
   不断迭代，让模型变得更好、更强、更快。  
   身为资料科学家，你应该尽可能将开发作业与系统整合外包或自动化，  
   让你能专注于：  
   - 洞察  
   - 推荐  
   - 设计  
   当你能清楚包装你的想法，它们就变成你在组织内的“价值交换货币”。

## 3.21 资料清理的四大要素：  
### Correcting（修正）、Completing（补全）、Creating（建立）、Converting（转换）

在此阶段，我们将透过以下四个步骤清理资料：  
1) 修正异常值与离群点  
2) 补全缺失资料  
3) 建立新特征  
4) 将字段转换为适合模型计算的格式  

---

### 1. **Correcting（修正）**  
检查资料后，目前看起来没有明显异常或不可接受的输入值。  
虽然 *Age* 与 *Fare* 可能包含离群值，但它们皆属合理范围，因此我们会等到完成探索性分析（EDA）后，再判断是否应排除。  

如果遇到完全不合理的数值，如 `age = 800`（应为 80），此时就可立即修正。  
但需注意：**修改原始资料需谨慎，以免破坏模型的真实性。**

---

### 2. **Completing（补全）**  
*Age*、*Cabin*、*Embarked* 字段存在缺失值。  
缺失资料会让模型产生偏误，因此在建模前必须先处理。

处理方式常见两种：  
- **删除纪录**（不建议，除非该笔资料确实不可用）  
- **填补资料（imputation）**（较佳作法）

推荐方法：  
- 类别型资料：填补“众数（mode）”  
- 数值型资料：填补“平均／中位数／平均＋随机标准差”  

进阶方法可依条件填补，如：  
- 依 Pclass 的平均年龄填补 Age  
- 依 Fare 与 SES 推估 Embarked  

但在采用复杂方法前，应先确定其确实能改善基本模型的效能。

本资料集采用：  
- Age → 以中位数填补  
- Cabin → 缺失太多，整栏舍弃  
- Embarked → 以众数填补  

（后续模型迭代时可再调整，以观察是否提高准确率）

---

### 3. **Creating（建立）**  
特征工程（Feature Engineering）指使用现有资料来创造新特征，以提升模型能力。

在本资料集中，我们将从姓名字段中萃取“称谓 Title”作为新特征，以检查其是否影响生还率。

---

### 4. **Converting（转换）**  
最后是资料格式的处理。  
本资料集中虽无日期或货币格式，但有“资料型态”问题。

部分字段（如 Sex、Embarked）以 *object* 型态汇入，不适合直接进行数学运算，因此需要转换。

在本资料集中，我们会将名目型字段转换成：

```python
# 补全：在训练与测试／验证资料中填补或删除缺失值
for dataset in data_cleaner:    
    # 使用中位数填补缺失的年龄
    dataset['Age'].fillna(dataset['Age'].median(), inplace = True)

    # 使用众数填补缺失的登船港口
    dataset['Embarked'].fillna(dataset['Embarked'].mode()[0], inplace = True)

    # 使用中位数填补缺失的票价
    dataset['Fare'].fillna(dataset['Fare'].median(), inplace = True)
    
# 删除 Cabin、Ticket 与 PassengerId 字段
drop_column = ['PassengerId','Cabin', 'Ticket']
data1.drop(drop_column, axis=1, inplace = True)

print(data1.isnull().sum())
print(data_val.isnull().sum())
```

```python
# 建立：为训练与测试／验证资料进行特征工程
for dataset in data_cleaner:    

    # 建立离散型特征：家庭人数
    dataset['FamilySize'] = dataset ['SibSp'] + dataset['Parch'] + 1

    # 建立是否独自旅行特征，预设为 1（独自）
    dataset['IsAlone'] = 1
    # 若家庭人数 > 1，则为 0（非独自）
    dataset['IsAlone'].loc[dataset['FamilySize'] > 1] = 0

    # 从姓名字段中萃取称谓（Title）
    dataset['Title'] = dataset['Name'].str.split(", ", expand=True)[1].str.split(".", expand=True)[0]

    # 使用 qcut 将票价分成四个区间
    dataset['FareBin'] = pd.qcut(dataset['Fare'], 4)

    # 使用 cut 将年龄分成五个区间
    dataset['AgeBin'] = pd.cut(dataset['Age'].astype(int), 5)


# 再次预览资料
data1.info()
data_val.info()
data1.sample(10)
```

## 3.23 格式转换（Convert Formats）

我们将把类别型资料转换为可供数学分析的“虚拟变数（dummy variables）”。  
编码类别变数的方法有很多种，本专案将使用 **sklearn** 和 **pandas** 的功能进行转换。

在此步骤中，我们也会定义资料建模所需的：  
- **x（自变数／特征／解释变数／预测变数）**  
- **y（依变数／目标变数／输出变数）**

---

### 相关文件（Source Documentation）：

* [Categorical Encoding](http://pbpython.com/categorical-encoding.html)
* [Sklearn LabelEncoder](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html)
* [Sklearn OneHotEncoder](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html)
* [Pandas Categorical dtype](https://pandas.pydata.org/pandas-docs/stable/categorical.html)
* [pandas.get_dummies](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.get_dummies.html)


```python
# 转换：使用 Label Encoder 将类别型字段转成数值（适用于训练与测试资料）
label = LabelEncoder()

for dataset in data_cleaner:    
    dataset['Sex_Code'] = label.fit_transform(dataset['Sex'])
    dataset['Embarked_Code'] = label.fit_transform(dataset['Embarked'])
    dataset['Title_Code'] = label.fit_transform(dataset['Title'])
    dataset['AgeBin_Code'] = label.fit_transform(dataset['AgeBin'])
    dataset['FareBin_Code'] = label.fit_transform(dataset['FareBin'])


# 定义 y 变数（目标／预测结果）
Target = ['Survived']

# 定义 x 变数（使用原始变数，适合视觉化）
data1_x = ['Sex','Pclass', 'Embarked', 'Title','SibSp', 'Parch', 'Age', 'Fare']
# 定义 x 变数（将类别型变数转为编码后的数值，用于模型计算）
data1_x_calc = ['Sex_Code','Pclass', 'Embarked_Code', 'Title_Code','SibSp', 'Parch', 'Age', 'Fare']
data1_xy =  Target + data1_x
print('Original X Y: ', data1_xy, '\n')


# 定义 x 变数（使用分箱后的字段，避免连续变数）
data1_x_bin = ['Sex_Code','Pclass', 'Embarked_Code', 'Title_Code', 'FamilySize', 'AgeBin_Code', 'FareBin_Code']
data1_xy_bin = Target + data1_x_bin
print('Original w/Bin X Y: ', data1_xy_bin, '\n')


# 定义 x 与 y（使用原始字段的 dummy variables）
data1_dummy = pd.get_dummies(data1[data1_x_calc])
data1_dummy_x = data1_dummy.columns.tolist()
data1_dummy_xy = Target + data1_dummy_x
print('Dummy Coding X Y: ', data1_dummy_xy, '\n')


# 定义 x 与 y（使用分箱字段的 dummy variables）
data1_dummy_bin = pd.get_dummies(data1[data1_x_bin])
data1_dummy_x_bin = data1_dummy_bin.columns.tolist()
data1_dummy_xy_bin = Target + data1_dummy_x_bin
print('Dummy Coding X Y: ', data1_dummy_xy_bin, '\n')

data1_dummy_bin.head()
```

## 3.25 切分训练资料与测试资料（Split Training and Testing Data）

因此，我们会使用 **sklearn** 的函式将训练资料再切分成两部分，  
通常采用 **75% 训练 / 25% 测试** 的比例。

这么做非常重要，因为可以避免模型“过度拟合（overfitting）”：  
也就是模型在某个资料子集上表现很好，但无法准确应用到同一资料集中的其他子集。

我们必须确保模型在训练期间从未看过“测试子集”，  
这样模型才不会靠“记住答案”来作弊，而是能真正学会泛化能力。

本步骤将使用：  
- [sklearn train_test_split](http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html)  
或  
- [sklearn 交叉验证（cross validation）函式](http://scikit-learn.org/stable/modules/cross_validation.html#cross-validation)

来完成资料切分。


```python
# 使用预设参数切分训练与测试资料
train1_x_bin, test1_x_bin, train1_y_bin, test1_y_bin = model_selection.train_test_split(data1[data1_x_bin], data1[Target])
train1_dummy_x, test1_dummy_x, train1_dummy_y, test1_dummy_y = model_selection.train_test_split(data1_dummy[data1_dummy_x], data1[Target])
train1_dummy_x_bin, test1_dummy_x_bin, train1_dummy_y_bin, test1_dummy_y_bin = model_selection.train_test_split(data1_dummy_bin[data1_dummy_x_bin], data1[Target])

# 显示资料形状
print("Data Shape: {}".format(data1.shape))
print("Train1 Shape: {}".format(train1_x_bin.shape))
print("Test1 Shape: {}".format(test1_x_bin.shape))
```

## 确认使用的算法类型：分类（Classification）

在确定使用 **监督式学习＋分类模型** 之后，可以将选择的算法范围缩小到：

### 常见分类算法（scikit-learn）
* [Ensemble Methods（集成算法）](http://scikit-learn.org/stable/modules/classes.html#module-sklearn.ensemble)
* [Generalized Linear Models（GLM）](http://scikit-learn.org/stable/modules/classes.html#module-sklearn.linear_model)
* [Naive Bayes（朴素贝氏）](http://scikit-learn.org/stable/modules/classes.html#module-sklearn.naive_bayes)
* [Nearest Neighbors（KNN）](http://scikit-learn.org/stable/modules/classes.html#module-sklearn.neighbors)
* [Support Vector Machines（SVM）](http://scikit-learn.org/stable/modules/classes.html#module-sklearn.svm)
* [Decision Trees（决策树）](http://scikit-learn.org/stable/modules/classes.html#module-sklearn.tree)


```python
from sklearn import model_selection, ensemble, gaussian_process, linear_model, naive_bayes, neighbors, svm, tree, discriminant_analysis

# 机器学习算法（MLA）清单
MLA = [
    # 集成方法 Ensemble
    ensemble.AdaBoostClassifier(),
    ensemble.BaggingClassifier(),
    ensemble.ExtraTreesClassifier(),
    ensemble.GradientBoostingClassifier(),
    ensemble.RandomForestClassifier(n_estimators=100),

    # 高斯过程
    gaussian_process.GaussianProcessClassifier(),

    # 一般化线性模型（GLM）
    linear_model.LogisticRegressionCV(max_iter=500),
    linear_model.PassiveAggressiveClassifier(),
    linear_model.RidgeClassifierCV(),
    linear_model.SGDClassifier(),
    linear_model.Perceptron(),

    # 朴素贝氏
    naive_bayes.GaussianNB(),

    # 最近邻算法
    neighbors.KNeighborsClassifier(n_neighbors=3),

    # 支援向量机
    svm.SVC(probability=True),
    svm.LinearSVC(),

    # 决策树
    tree.DecisionTreeClassifier(),
    tree.ExtraTreeClassifier(),

    # 判别分析
    discriminant_analysis.LinearDiscriminantAnalysis(),
    discriminant_analysis.QuadraticDiscriminantAnalysis()
]

# 建立算法比较表
MLA_columns = ['MLA Name', 'MLA Parameters', 'MLA Train Accuracy', 'MLA Test Accuracy', 'MLA Time']
MLA_compare = pd.DataFrame(columns=MLA_columns)

Target = 'Survived'
y = data1[Target].values.ravel()

row_index = 0
for alg in MLA:
    print(f"Running: {alg.__class__.__name__}")

    MLA_compare.loc[row_index, 'MLA Name'] = alg.__class__.__name__
    MLA_compare.loc[row_index, 'MLA Parameters'] = str(alg.get_params())

    # 进行 10-fold 交叉验证（包含训练与测试分数）
    cv_results = model_selection.cross_validate(
        alg,
        data1[data1_x_bin],
        y,
        cv=10,
        return_train_score=True,
        n_jobs=-1
    )

    MLA_compare.loc[row_index, 'MLA Time'] = cv_results['fit_time'].mean()
    MLA_compare.loc[row_index, 'MLA Train Accuracy'] = cv_results['train_score'].mean()
    MLA_compare.loc[row_index, 'MLA Test Accuracy'] = cv_results['test_score'].mean()

    # 在完整资料集上重新训练模型（交叉验证之后）
    MLA[row_index] = alg.fit(data1[data1_x_bin], data1[Target])

    row_index += 1

# 排序显示结果（依测试准确率由高到低）
MLA_compare.sort_values(by=['MLA Test Accuracy'], ascending=False, inplace=True)
print(MLA_compare)
MLA_compare
```

## 手作决策树：规则说明

决策树的游戏规则：

- 若一个子群中 **50% 以上生还 → 全部设为生还（1）**
- 若一个子群中 **50% 以上死亡 → 全部设为死亡（0）**
- 若子群样本 < 10（约占资料 10%）→ 停止下探
- 若加入特征后准确率没有提升或下降 → 停止下探

## 5.11 使用交叉验证（Cross-Validation, CV）评估模型效能

在步骤 5 中，我们使用了  
[sklearn cross_validate](http://scikit-learn.org/stable/modules/cross_validation.html#multimetric-cross-validation)  
来训练、测试并评估模型效能。

记住一件事：  
为了评估模型是否具有“真正的预测能力”，我们**一定要**把资料分成：

- **训练资料（train）**：用来建立模型  
- **测试资料（test）**：用来评估模型

否则模型会 **overfitting（过度拟合）**：  
模型只会“记住”训练资料的答案，而不是学到可泛化的规则。  
这就像作业和小考都偷看答案拿 100 分，但到了正式考试却完全不会写。

---

## 为什么要用 Cross-Validation？

当资料集很小时，  
强行切割成训练资料与测试资料，  
会使模型损失大量可用讯息。

**交叉验证（CV）** 的做法是：

1. 把资料切成 k 份  
2. 每次取其中 1 份当测试资料，其余 k-1 份当训练资料  
3. 重复 k 次  
4. 取得所有测试结果的“平均值”当最终效能

这样可以：

- **充分利用所有资料**
- **避免单次切割造成的偏差**
- **得到更可靠的效能估计**

缺点是：  
计算量增加，需要更多电脑资源。

## 决策树（Decision Tree）优点  
资料来源：sklearn 官方文件  
<http://scikit-learn.org/stable/modules/tree.html#classification>

> **决策树的优点：**
* 易于理解与解释，并可视觉化。
* 几乎不需要资料前处理（不需正规化、不需 dummy 变数，虽然不能处理缺失值）。
* 预测成本低（时间复杂度与训练资料笔数呈对数关系）。
* 能同时处理数值与类别资料（其他算法通常只擅长某一种）。
* 能处理多输出问题（multi-output）。
* 属于“白箱模型”：逻辑透明，容易解释（相较于黑箱如神经网络）。
* 可用统计方法评估其可靠性。
* 即使资料不完全符合算法假设，也通常能维持良好表现。

---

## 决策树（Decision Tree）缺点

> **决策树的缺点：**
* 容易产生过度复杂的树，导致 **overfitting（过度拟合）**。  
  → 必须使用剪枝、限制最大深度、设定叶节点最小样本数等方法（sklearn 不支援传统剪枝）。
* 对资料非常敏感，小变动可能生成完全不同的树。  
  → 可透过集成方法（ensemble）改善。
* 找到最佳决策树是 NP-complete 问题，因此实务算法多为 greedy heuristics，  
  不能保证得到全域最佳的树。
* 某些模式难以用树表达（如 XOR、Parity、Multiplexer 等）。
* 若资料不平衡，决策树容易产生偏差。  
  → 建议在训练前先平衡资料。
